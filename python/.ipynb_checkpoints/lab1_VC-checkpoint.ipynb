```python
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import Video
import cv2
import os
```

### Tarea 1 -  Cargar el dataset i dividirlo en dos partes iguales, una para modelar el fondo y otra para detectar

Vamos a usar las imágenes de la carpeta input que van desde la _in001051.jpg_ hasta la _in001350.jpg. Las 150 primeras son para entrenar y las 150 restate

```python
dataset_base_path = '../data/'
images_size = (240, 320) # Esto lo he visto "a mano", pero viene bien la variable
```

```python
first_image_n = 1051
last_image_n = 1350

all_images = np.empty((images_size[0], images_size[1], 300), np.float64) # Matriz con 300 "capas" para guardar todos los frames

for i in range(first_image_n, last_image_n+1):
    image_path = f'{dataset_base_path}input/in00{i}.jpg'
    all_images[:,:,i-first_image_n] = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

Mostramos una imagen cualquiera para ver que todo se han cargado correctamente en la matriz:

```python
plt.imshow(all_images[:,:,200], 'gray');
```

Ahora separamos el dataset en test y train:

```python
train_images = all_images[:,:,0:150]
test_images = all_images[:,:,150:]
```

Comprobamos que las medidas de estas particiones tienen sentido y coinciden con la pare del dataset que queremos:

```python
print(train_images.shape)
print(test_images.shape)
print(np.array_equal(all_images[:,:,0:150], train_images))
print(np.array_equal(all_images[:,:,150:], test_images))
```

### Tarea 2 - Calcular la media y la desviación estándar

Queremos obtener dos matrices. La primera será una que contenga el valor medio de cada pixel de las imágenes de _train_ y la segunda será la desviación estándar de cada uno de los píxeles de las imágenes de _train_. Por lo tanto esperamos 2 matrices com medidas 240x320:

```python
training_mean = np.sum(train_images, axis=2)/train_images.shape[2]
```

Visualizamos que el resultado obtenido tiene sentido:

```python
plt.imshow(training_mean, 'gray');
```

```python
training_std_dev = np.sqrt(np.sum(np.power(train_images-training_mean.reshape(training_mean.shape[0],training_mean.shape[1],1),2),axis=2)/train_images.shape[2])
```

Visualizamos que el resultado obtenido tiene sentido:

```python
plt.imshow(training_std_dev, 'gray');
```

### Tarea 3 - Segmentar los coches restando el modelo de fondo

Restamos la media a los _frames_ de _test_ para "eliminar" las partes sin movimiento para intentar detectar los coches: 

```python
subtracted = test_images-training_mean.reshape(training_mean.shape[0],training_mean.shape[1],1)
subtracted = subtracted < 0.0001
```

Comprobamos que el resultado obtenido tiene sentido:

```python
plt.imshow(subtracted[:,:,20], 'gray');
```

### Tarea 4 - Segmentar los coches con un modelo más elaborado

Utilizaremos una fórmula más compleja para calcular si los píxeles pertenecen al fondo o a objetos en movimiento. Dentro de esta fórmula, tenemos dos parámetros α y β que podemos variar para encontrar la perfecta conjunción entre ambos.

```python
alfa = -0.2
beta = -20
subtracted = test_images-training_mean.reshape(training_mean.shape[0],training_mean.shape[1],1)
subtracted = subtracted < (alfa * training_std_dev.reshape(training_std_dev.shape[0],training_std_dev.shape[1],1) + beta)
subtractedAux = subtracted #guardamos una copia para las métricas
```

Comprobamos que el resultado obtenido tiene sentido:

```python
plt.imshow(subtracted[:,:,20], 'gray');
```

Vemos una mejora substancial donde el ruido causado por el movimiento de las hojas de los árboles se ha eliminado en gran medida mientras que el movimiento de los coches es lo más resaltado.

### Tarea 5 - Grabar un vídeo con los resultados

Usando los resultados obtenidos con las tareas anteriores, generaremos una secuencia de vídeo mostrando todos los frames.

```python
video_path = "../data/videos/"
video_name = "VideoSample.avi"

def make_video(option):
            
    for f in os.listdir(video_path):
        os.remove(os.path.join(video_path, f))

    subtracted = subtractedAux
    enhancedOption = option

    kernel = np.ones((5,5),np.uint8)

    try:
        os.stat(video_path)
    except:
        os.mkdir(video_path)
    subtracted = subtracted.astype(np.uint8)
    for frame in range(test_images.shape[2]):
        if enhancedOption == "erode":
            subtracted[:,:,frame] = cv2.erode(subtracted[:,:,frame], kernel)
        elif enhancedOption == "dilate":
            subtracted[:,:,frame] = cv2.dilate(subtracted[:,:,frame], kernel)
        elif enhancedOption == "opening":
            subtracted[:,:,frame] = cv2.morphologyEx(subtracted[:,:,frame], cv2.MORPH_OPEN, kernel)
        elif enhancedOption == "closing":
            subtracted[:,:,frame] = cv2.morphologyEx(subtracted[:,:,frame], cv2.MORPH_CLOSE, kernel)
        elif enhancedOption == "custom":
            subtracted[:,:,frame] = cv2.erode(subtracted[:,:,frame], kernel, iterations=1)
            subtracted[:,:,frame] = cv2.dilate(subtracted[:,:,frame], kernel, iterations=5)

        cv2.imwrite((f"{video_path}/{frame}.png"),subtracted[:,:,frame]*255)


    frames = [img for img in os.listdir(video_path)]
    sorted_frames = sorted(frames, key=lambda frame:int(frame.split(".")[0]))
    #print(sorted_frames)
    height, width, layers = cv2.imread(os.path.join(video_path, sorted_frames[0])).shape

    final_video_path = "../data/"
    fps = 30
    video = cv2.VideoWriter(final_video_path+enhancedOption+video_name, cv2.VideoWriter_fourcc(*"DIVX"), fps, (width, height))
    cv2.VideoWriter()

    for frame in sorted_frames:
        video.write(cv2.imread(os.path.join(video_path, frame)))

    cv2.destroyAllWindows()
    video.release()
```

Veremos ahora el resultado de aplicar las mejoras sobre los distintos fotogramas del video.

```python
plt.imshow(subtracted[:,:,20], 'gray');
```

### Tarea 6 - Evaluar la bondad de los resultados

Para evaluar cuan bueno es nuestro modelo frente el "groundtruth" dado en el mismo dataset donde encontramos los fotogramas de la cámara de tráfico, utilizaremos métricas como el "accuracy". Para ello, compararemos fotograma a fotograma los resultados de diferentes configuraciones aplicadas a nuestro programa contra el "groundtruth" antes mencionado.

También acompañaremos el "accuracy" con otras métricas tales como "precision", "recall" i "f1-score", que nos ayudarán a tener una mejor perspectiva de la calidad de los resultados obtenidos.

```python
#Accuracy = TP+TN / TP+TN+FP+FN

groundtruth = os.listdir("../data/groundtruth/")
groundtruthDataset = groundtruth[1201:1351]

optionset = ["simple", "erode", "dilate", "opening", "closing", "custom"]
#optionset = ["custom"]

print("--------------------------------------")

for option in optionset:
    
    acc = []
    prec = []
    rec = []
    f1 = []
    
    make_video(option)
    for i in range(len(groundtruthDataset)):
        image_path = f'../data/groundtruth/gt00{i+1201}.png'
        video_frame_path = f'../data/videos/{i}.png'
        matGT = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        matTest = cv2.imread(video_frame_path, cv2.IMREAD_GRAYSCALE)
        matTest = matTest.astype(np.uint8)

        mat0 = np.where(matTest == 0)
        res0 = matTest[mat0] - matGT[mat0]
        mat1 = np.where(matTest != 0)
        res1 = matTest[mat1] - matGT[mat1]

        TP = res1.size - np.count_nonzero(res1)
        TN = res0.size - np.count_nonzero(res0)
        FP = np.count_nonzero(res1)
        FN = np.count_nonzero(res0)

        acc.append((TP+TN)/(TP+TN+FP+FN))
        precAct = TP/(TP+FP)
        prec.append(precAct)
        recAct = TP/(TP+FN)
        rec.append(recAct)
        f1.append((2*(recAct*precAct))/(recAct+precAct))

    print(f"Results on {option}")
    print("Accuracy: ", np.median(acc)*100, "%")
    print("Precision: ", np.median(prec)*100, "%")
    print("Recall: ", np.median(rec)*100, "%")
    print("F1Score: ", np.median(f1)*100, "%")
    print("--------------------------------------")
    
```

Vemos en los resultados que todos tienen un nivel muy alto de "accuracy", pero unos métodos tienen mejores puntajes en ciertas métricas. Esto se debe a que al utilizar distintas mecánicas, quitar area del objeto como puede ser el método "erode" consigue más precisión pero pierde mucho "recall" en relación al "dilate", que es el método inverso.

En general, todos los métodos serían válidos ya que dan unos resultados muy aceptables, pero creemos que el método "closing" por los resultados según las métricas i el "custom" por las formas más faciles de tratar posteriormente, son las mejores propuestas para este problema en concreto.
